# Visualisasi Data

```python
sns.set(rc={'figure.figsize':(20,7)})
```

```python
def show_top_n_barplot(n:int, column_name: str, x_axis_col: str, x_label: str, y_label: str, title: str):
    # top 10 songs based on streaming.
    sns.barplot(data=spotify_df.sort_values(column_name, ascending=False).head(n),x=x_axis_col,y=column_name)
    plt.xticks(rotation=45, ha="right") 
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()
```

```python
# Top 10 Most Streamed Songs
n = 10
column_name = 'streams'
x_axis_col = 'track_name'
x_label = 'Song / Track Name'
y_label = 'Number of Streams'
title = f'Top {n} Most Streamed Songs'
show_top_n_barplot(n, column_name, x_axis_col, x_label, y_label, title)
```

```python
# Top 5 Most Songs Across All Playlists
n = 5
column_name = 'in_spotify_playlists'
x_axis_col = 'track_name'
x_label = 'Song / Track Name'
y_label = 'Total Playlist'
title = f'Top {n} Most Songs Across All Playlists'
show_top_n_barplot(n, column_name, x_axis_col, x_label, y_label, title)
```

```python
#visualisasi distribusi data streams

sns.set(style="whitegrid")
sns.histplot(spotify_df['streams'], kde=True, color='blue')

plt.xlabel('Streams')
plt.ylabel('Frekuensi')
plt.title('Distribusi data streams')

plt.show()
```

```python
# Buat boxplot untuk kolom "streams"

sns.boxplot(data = spotify_df, x = 'streams')

# Tampilkan boxplot
plt.show()
```

```python
print("Statistik Deskriptif untuk data kolom streams")
rata_rata = spotify_df['streams'].mean()
min_value = spotify_df['streams'].min()
median = spotify_df['streams'].median()
q1 = spotify_df['streams'].quantile(0.25)
q3 = spotify_df['streams'].quantile(0.75)
max_value = spotify_df['streams'].max()
standar_deviasi = spotify_df['streams'].std()
print(f'Rata-rata jumlah stream : {rata_rata}')
print(f'Nilai Minimum : {min_value}')
print(f'Median (Q2): {median}')
print(f'Quartil 1 : {q1}')
print(f'Quartil 3 : {q3}')
print(f'Nilai Maksimum : {max_value}')
print(f'standar Deviasi : {standar_deviasi}')
```

Kesimpulan yang bisa diambil dari visualisasi data streams :
1. Kebanyakan top lagu spotify yang memiliki streams terbanyak memiliki jumlah streams tersebar dari sekitar 100 juta sampai 600 juta.
2. Dataset top lagu Spotify ini memiliki jumlah streams terkecil yaitu 2762 dan jumlah streams terbanyak yaitu 3562543890 (3.5 miliar).
3. Dilihat dari nilai standar deviasi yang lebih besar dari nilai rata-rata menandakan bahwa data streams sangat bervariasi.
4. Dilihat dari boxplot data streams, semua pencilan tersebar dengan jumlah streams lebih dari quartil 3
