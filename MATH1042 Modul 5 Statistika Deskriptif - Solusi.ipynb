## 5.1 Manipulasi Data CSV dengan menggunakan Pandas

Struktur data utama yang digunakan dalam Pandas:
* ```DataFrame```, yaitu suatu tabel dengan kolom-kolom berperan sebagai atribut data dan baris-baris berperan sebagai data record.

````python
# Impor Data ke dalam struktur data Pandas

# To do:
# 1. Import library pandas
# 2. Import file CSV menjadi suatu dataframe di Pandas
# 3. Tampilkan file CSV dalam format dataframe pandas

import pandas as pd
df = pd.read_csv('https://calvin.instructure.com/courses/595/files/96586/download?download_frd=1')
display(df.head(10))
````

## 5.2 Visualisasi Data Diskrit/Kategorial

Dalam bagian ini, kita akan memvisualisasikan sebaran data ```"Highest_Qualified_Member"``` dan ```No_of_Earning_Members```, yang merupakan data kategorial dan diskrit, secara berturut-turut.

````python
# Membangun CountTable/Tabel Frekuensi

# panggil Series untuk kolom "Highest_Qualified_Member"
academic_series = df["Highest_Qualified_Member"]
# display(academic_series)

# konversi menjadi tabel frekuensi
freq_table = academic_series.value_counts()
display(freq_table)
````

````python
# Membangun Pie Chart dengan Matplotlib

# import library matplotlib
from matplotlib import pyplot as plt

# mendefinisikan kanvas untuk menggambar diagram
fig = plt.figure(figsize =(10, 5)) 
# figsize = (a, b) artinya a ukuran lebar horizontal, b ukuran panjang vertikal

# perintah untuk menggambar diagram di kanvas
plt.pie(freq_table, labels = freq_table.index, autopct='%1.1f%%')

# perintah untuk memberi judul diagram
plt.title("Highest Academic Degree of Household Members")

# perintah untuk menampilkan gambar di kanvas
plt.show()
````

````python
# Membangun Bar Chart dengan Matplotlib

# Membangun list nilai x dan y untuk digambar di bar chart
x = freq_table.index.to_list()
y = freq_table.to_list()

print(x)
print(y)

# mendefinisikan kanvas untuk menggambar diagram
fig = plt.figure(figsize =(8, 5)) 

# perintah untuk menggambar diagram di kanvas
plt.bar(x, y, color='grey')

# perintah untuk memberi label sumbu x dan y
plt.xlabel("Academic Degree")
plt.ylabel("Frequency")

# perintah untuk memberi judul diagram
plt.title("Highest Academic Degree of Household Members")

# perintah untuk menampilkan gambar di kanvas
plt.show()
````

### Highlights:

* Kelompok mayoritas dalam kumpulan data tersebut adalah yang memiliki gelar akademik "Graduate", yaitu sebesar 38%.
* Kelompok "Post-Graduate" dan "Illiterate" merupakan dua kelompok yang paling sedikit perwakilannya, secara berturut-turut mewakili 12% dan 10% dari keseluruhan data.
* Dua kelompok lainnya memiliki keterwakilan yang sama-sama sebesar 20%, yaitu "Professional" dan "Under-Graduate".

## 5.3 Visualisasi data Kontinu

Dalam bagian ini, kita akan memvisualisasikan sebaran data ```Mthly_HH_Income``` dan ```Mthly_HH_Expense```, yang keduanya merupakan data kontinu.

````python
# Membangun histogram dari Pandas


# Membangun kanvas matplotlib
fig = plt.figure(figsize =(7, 5)) 

# Series Pandas memiliki method untuk langsung menggambar histogram di kanvas matplotlib
df["Mthly_HH_Income"].hist()

# perintah untuk memberi label sumbu x dan y
plt.xlabel("Amount of Income")
plt.ylabel("Frequency")

# perintah untuk memberi judul diagram
plt.title("Distribution of Incomes among Households")

# perintah untuk menampilkan gambar di kanvas
plt.show()
````

````python
# Alternatif lain menggunakan Seaborn
# Referensi: https://www.sharpsightlabs.com/blog/seaborn-histogram/ 

# import library seaborn
import seaborn as sns

# untuk menggunakan style khas Seaborn. Referensi: https://seaborn.pydata.org/tutorial/aesthetics.html
sns.set_style("whitegrid")

# seaborn dapat langsung menangkap data dari dataframe
_=sns.histplot(data=df, x='Mthly_HH_Income')

plt.title("Distribution of Incomes among Households")
plt.show()
````

````python
# Modifikasi jumlah bins


def plot_hist_with_num_bins(df, num_bins):
    # perintah ini meminta sns untuk membuat histogram dengan 'num_bins' buah bin
    _=sns.histplot(data=df, x='Mthly_HH_Income', bins = num_bins)
    plt.title(f"Distribution of Incomes among Households with {num_bins} bins")
    plt.show()
    plt.close()

plot_hist_with_num_bins(df, 4)
plot_hist_with_num_bins(df, 6)
````

````python
# Modifikasi ukuran bins


def plot_hist_with_bin_size(df, bin_size):
    # perintah ini meminta sns untuk membuat histogram dengan ukuran bin sebesar bin_size
    _=sns.histplot(data=df, x='Mthly_HH_Income', binwidth = bin_size)
    plt.title(f"Distribution of Incomes among Households with bins size of {bin_size}")
    plt.show()
    plt.close()

plot_hist_with_bin_size(df, 10000)
plot_hist_with_bin_size(df, 5000)
````

````python
# Membangun boxplot

fig = plt.figure(figsize =(5, 7)) 
sns.set_style("white")

# perintah di bawah ini meminta sns untuk membuat boxplot dengan arah vertikal (sumbu y)
_=sns.boxplot(data = df, y = 'Mthly_HH_Income') # coba ganti y menjadi x dalam perintah tsb

plt.title("Boxplot of Incomes among Households")
plt.show()
````

### Highlights:

Terkait data ```"Mthly_HH_Income"```,

### 5.3.1 Kaitan antara dua fitur kontinu

````python
# Melihat kaitan antara Income dan Expense dengan histogram multivariabel

sns.set_theme()

_=sns.histplot(data = df, x = "Mthly_HH_Income", color = "blue", label = "Income", alpha=0.5)
_=sns.histplot(data = df, x = "Mthly_HH_Expense", color = "maroon", label = "Expense", alpha=0.5)

plt.legend()
plt.xlabel("Amount")
plt.title("Histogram of Monthly Income vs Expense")
plt.show()
````

````python
# Melihat kaitan antara Income dan Expense dengan scatter plot

# Dengan Matplotlib

# membangun vektor x dan y untuk titik-titik yang akan diplot
x = df["Mthly_HH_Income"].to_list()
y = df["Mthly_HH_Expense"].to_list()


fig = plt.figure(figsize =(7, 5))

# menggambar scatter plot di kanvas matplotlib
plt.scatter(x, y, marker = 'x')

plt.xlabel("Income")
plt.ylabel("Expense")
plt.title("Distribution of Income vs Expense")
plt.show()
````

````python
# Melihat kaitan antara Income dan Expense dengan scatter plot

# Dengan Seaborn

fig = plt.figure(figsize =(7, 5))

# menggambar scatter plot dengan sns langsung dengan df
sns.scatterplot(data = df, x="Mthly_HH_Income", y="Mthly_HH_Expense")

plt.xlabel("Income")
plt.ylabel("Expense")
plt.title("Distribution of Income vs Expense")
plt.show()
````

````python
# Ada berapa banyak yang expense >= income? Tampilkan dengan visualisasi

# beri warna merah untuk titik yang expense >= income & biru untuk lainnya
x_red, y_red = [], []
x_blue, y_blue = [], []

x = df["Mthly_HH_Income"].to_list()
y = df["Mthly_HH_Expense"].to_list()

for i in range(len(x)):
    if x[i]<=y[i]: # income < expense
        x_red.append(x[i])
        y_red.append(y[i])
    else:
        x_blue.append(x[i])
        y_blue.append(y[i])

fig = plt.figure(figsize =(7, 5))

# menggambar scatter plot yang dibedakan warna titiknya
plt.scatter(x_red, y_red, color="red", label="expense >= income")
plt.scatter(x_blue, y_blue, color="blue", label="expense < income")

plt.xlabel("Income")
plt.ylabel("Expense")
plt.title("Distribution of Income vs Expense")
plt.legend()
plt.show()
````

### Highlights:

* Dari histogram dapat terlihat memang secara umum sebaran data expense lebih memiliki nilai yang lebih rendah daripada sebaran data income.
* Dari scatter plot, dapat terlihat memang hanya sebagian kecil saja keluarga yang memiliki masalah expense yang lebih besar daripada income, sebagaimana ditandai oleh titik berwarna merah.

### 5.3.2 Kaitan antara variabel kontinu dan variabel diskrit/kategorial

```"Highest_Qualified_Member"``` adalah variabel kategorial dan ```"No_of_Earning_Members"``` adalah variabel diskrit. Kita akan mengaitkan nilai Income dan Expense raltif terhadap kedua variabel tersebut.

````python
# Pandas memiliki fitur untuk melakukan grup terhadap data kategorial/diskrit

# df_group adalah objek untuk df yang digrup berdasarkan fitur tertentu
df_group = df.groupby("Highest_Qualified_Member")

# perintah untuk menampilkan semua data yang terpisahkan berdasarkan grupnya.
for name, group in df_group:
   print(f"Showing dataframe for Group \"{name}\"...")
   display(group)
````

````python
# menampilkan rataan data numerik dari setiap kelompok grup
display(df_group.mean())

# menampilkan standar deviasi data numerik dari setiap kelompok grup
display(df_group.std())
````

````python
# membangun diagram batang rataan & stdev income per kelompok academic

# membuat list berisi rataan, stdev income
income_ave = df_group.mean()["Mthly_HH_Income"]
income_std = df_group.std()["Mthly_HH_Income"]
group = df_group.mean().index

sns.set_theme()
fig = plt.figure(figsize =(7, 5))

plt.bar(group, income_ave, color="skyblue")
# errorbar digambar untuk memvisualisasikan standar deviasi
plt.errorbar(group, income_ave, yerr = income_std, fmt="o", color="blue")

plt.title("Average & Stdev Income relative to Academic Degree")
plt.show()
plt.close()
````

````python
# Membuat histogram nilai "Mthly_HH_Income" per kelompok "Highest_Qualified_Member"

sns.set_style("whitegrid")

fig = plt.figure(figsize =(10, 5))

# spesifikasikan kelompok yang hendak dibedakan warna histogramnya pada bagian hue
_ = sns.histplot(data = df, x = "Mthly_HH_Income", hue = "Highest_Qualified_Member", bins = 10)

plt.title("Histogram of Incomes grouped by Academic Qualification")
plt.show()
````

````python
# Bangun histogram setiap kelompoknya satu demi satu

for name, group in df_group:
    fig = plt.figure(figsize =(10, 5))

    _ = sns.histplot(data = group, x = "Mthly_HH_Income", binwidth = 10000)

    plt.title(f"Histogram of Incomes of {name}")
    
    #perintah untuk membuat batasan nilai sumbu x dan y yang hendak ditampilkan
    plt.ylim(0,6.5)
    plt.xlim(0,100000)
    plt.show()
    plt.close()
````
